```python
import matplotlib.pyplot as plt
```

```python
def plot_losses(train_loss_file, test_loss_file, ylim):
    train_f = open(train_loss_file, "r")
    test_f = open(test_loss_file, "r")

    train_loss_raw = train_f.readlines()
    test_loss_raw = test_f.readlines()
    train_loss = []
    test_loss = []

    for i in range(len(train_loss_raw)):
        train_loss.append(float(train_loss_raw[i]))
        test_loss.append(float(test_loss_raw[i]))
    plt.axis([0, len(train_loss), 0, ylim])
    plt.plot(train_loss, label="training loss")
    plt.plot(test_loss, label="testing loss")
    plt.legend()
    train_f.close()
    test_f.close()
```

## DeepSet withd deck2vec Learning curve

```python
plot_losses("train_loss_deepset.txt", "test_loss_deepset.txt", 10)
```
